<h1>Zadanie 1<h1>

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, io

img = data.camera()
io.imshow(data.camera())
```

```python
ay, ax = np.histogram(img, range=(0, 256), bins=256)

plt.plot(ax[:-1], ay, color='black')

plt.title('histogram')
ax = plt.xlabel('jasność')
ay = plt.ylabel('piksele')

plt.show()
```

```python
#ax = plt.hist(data.camera().ravel(), bins = 8, color = 'Blue', rwidth=0.85, range=(0,256))
#ax = plt.xlabel('jasność')
#ay = plt.ylabel('piksele')
#plt.show()

ay, ax = np.histogram(data.camera(), bins = 8, range = (0, 256))

x2str = [str(int(i)) for i in ax]
plt.bar(x2str[:-1], ay)

plt.title('Histogram dla 8 kubełków')
ax = plt.xlabel('jasność')
ay = plt.ylabel('piksele')
```

```python
ay, ax = np.histogram(img, range=(0, 256), bins=256)

f = np.cumsum(ay) * 0.000004
#print(f)

plt.plot(ax[:-1], f, color="black")

plt.title("Histogram skumulowany")
ax = plt.xlabel("jasność")
ay = plt.ylabel("funkcja prawdopodobieństwa")
```

```python
dark_img = io.imread('dark_image.png')
io.imshow(dark_img)
```

```python
dark_ay, dark_ax = np.histogram(dark_img, range=(0, 256), bins=256)

plt.plot(dark_ax[:-1], dark_ay, color='black')

plt.title('histogram')
dark_ax = plt.xlabel('jasność')
dark_ay = plt.ylabel('piksele')

plt.show()
```

```python
import skimage

dark_img2 = skimage.exposure.rescale_intensity(dark_img, (5, 220))
dark_img2 = skimage.exposure.adjust_gamma(dark_img2, 0.7, 1)
io.imshow(dark_img2)
```

```python
dark_y, dark_x = np.histogram(dark_img2, range=(0,256), bins=256)

plt.plot(dark_x[:-1], dark_y, color="black")

plt.title('histogram')
dark_x = plt.xlabel('jasność')
dark_y = plt.ylabel('piksele')

plt.show()
```
