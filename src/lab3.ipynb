<h1> Zadanie 1 <h1>

```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu, threshold_mean, threshold_yen, threshold_triangle

image1 = io.imread("gears1.png")
ay, ax = np.histogram(image1, range=(0, 256), bins=256)
io.imshow(image1)
```

```python
thresh_mean_1 = threshold_mean(image1)
image1_seg_mean = (image1 < thresh_mean_1) * 1

fig, ax = plt.subplots(ncols=2, figsize=(12, 6))
ax[0].imshow(image1_seg_mean, cmap=plt.cm.gray)

ax[1].plot(ay, color="black")
ax[1].axvline(thresh_mean_1, color="r")
ax[1].set_title("histogram")
ax[1].set_xlabel("jasno")
ax[1].set_ylabel("piksele")
```

```python
thresh_otsu_1 = threshold_otsu(image1)
image1_seg_otsu = (image1 < thresh_otsu_1) * 1

fig, ax = plt.subplots(ncols=2, figsize=(12, 6))
ax[0].imshow(image1_seg_otsu, cmap=plt.cm.gray)

ax[1].plot(ay, color="black")
ax[1].axvline(thresh_otsu_1, color="r")
ax[1].set_title("histogram")
ax[1].set_xlabel("jasno")
ax[1].set_ylabel("piksele")
```

<h1>Zadanie 2 <h1>

```python
from skimage.filters import try_all_threshold

image2 = io.imread('brain_tumor.png')

fig, ax = try_all_threshold(image2, figsize=(10, 8), verbose=False)
plt.show()
```

```python
ay, ax = np.histogram(image2, range=(0, 256), bins=256)
io.imshow(image2)
```

```python
thresh_yen_2 = threshold_yen(image2)
image2_seg_yen = (image2 > thresh_yen_2) * 1

fig, ax = plt.subplots(ncols=2, figsize=(12, 6))
ax[0].imshow(image2_seg_yen, cmap=plt.cm.gray)

ax[1].plot(ay, color="black")
ax[1].axvline(thresh_yen_2, color="r")
ax[1].set_title("histogram")
ax[1].set_xlabel("jasno")
ax[1].set_ylabel("piksele")
```

```python
thresh2 = 226
image2_seg = (image2 > thresh2) * 1

fig, ax = plt.subplots(ncols=2, figsize=(12, 6))
ax[0].imshow(image2_seg, cmap=plt.cm.gray)

ax[1].plot(ay, color="black")
ax[1].axvline(thresh2, color="r")
ax[1].set_title("histogram")
ax[1].set_xlabel("jasno")
ax[1].set_ylabel("piksele")
```

<h1>Zadanie 3<h1>

```python
image3 = io.imread('printed_text.png')

fig, ax = try_all_threshold(image3, figsize=(10, 8), verbose=False)
plt.show()
```

```python
io.imshow(image3)
ay, ax = np.histogram(image3, range=(0, 256), bins=256)
```

```python
thresh_triangle_3 = threshold_triangle(image3)
image3_seg_triangle = (image3 > thresh_triangle_3) * 1

fig, ax = plt.subplots(ncols=2, figsize=(12, 6))
ax[0].imshow(image3_seg_triangle, cmap=plt.cm.gray)

ax[1].plot(ay, color="black")
ax[1].axvline(thresh_triangle_3, color="r")
ax[1].set_title("histogram")
ax[1].set_xlabel("jasno")
ax[1].set_ylabel("piksele")
```

```python
thresh3 = 50
image3_seg = (image3 > thresh3) * 1

fig, ax = plt.subplots(ncols=2, figsize=(12, 6))
ax[0].imshow(image3_seg, cmap=plt.cm.gray)

ax[1].plot(ay, color="black")
ax[1].axvline(thresh3, color="r")
ax[1].set_title("histogram")
ax[1].set_xlabel("jasno")
ax[1].set_ylabel("piksele")
```

```python
from skimage.filters.rank import maximum
from skimage.morphology import disk

image3_background = maximum(image3, disk(5))
io.imshow(image3_background)
```

```python
image3_out = image3_background - image3
io.imshow(image3_out)
```

```python
fig, ax = try_all_threshold(image3_out, figsize=(10, 8), verbose=False)
plt.show()
```

```python
thresh_mean_out_3 = threshold_mean(image3_out)
image3_out_seg_mean = (image3_out > thresh_mean_out_3) * 1

fig, ax = plt.subplots(ncols=2, figsize=(12, 6))
ax[0].imshow(image3_out_seg_mean, cmap=plt.cm.gray)

ax[1].plot(ay, color="black")
ax[1].axvline(thresh_mean_out_3, color="r")
ax[1].set_title("histogram")
ax[1].set_xlabel("jasno")
ax[1].set_ylabel("piksele")
```

```python
io.imshow(image3_out_seg_mean, cmap=plt.cm.gray)
io.imsave("image3_out.jpg", image3_out_seg_mean)
```
