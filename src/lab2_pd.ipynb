```python
from skimage import data, io
import numpy as np
import matplotlib.pyplot as plt

img = data.chelsea()
#io.imshow(img)

colors = ("red", "green", "blue")
channel = (0, 1, 2)

plt.xlim([0, 256])
for channel_id, c in zip(channel, colors):
    ay, ax = np.histogram(img[:, :, channel_id], bins=256, range=(0, 256))
    plt.plot(ax[0:-1], ay, color=c)

#plt.xlabel("Wartość koloru")
#plt.ylabel("Pixele")
plt.legend(['czerwony', 'zielony', 'niebieski'])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

y_value_red, x_value = np.histogram(img[:, :, 0], bins=8, range=(0, 256))
y_value_green, x_value = np.histogram(img[:, :, 1], bins=8, range=(0, 256))
y_value_blue, x_value = np.histogram(img[:, :, 2], bins=8, range=(0, 256))

x2str = [str(int(i)) for i in x_value]
x = np.arange(len(x2str) - 1)
bar_width = 0.25

b1 = ax.bar(x - bar_width, y_value_red, width=bar_width, color="red")
b2 = ax.bar(x, y_value_green, width=bar_width, color="green")
b3 = ax.bar(x + bar_width, y_value_blue, width=bar_width, color="blue")

# print(x2str)
# print(x)
labels = []
for i in range(0, len(x)):
    labels.append("(" + x2str[i] + "," + x2str[i + 1] + ")")
# print(labels)

ax.set_ylabel("Pixele")
ax.set_xlabel("Zakresy")
ax.set_title("Histogram kanalow kolorow")
ax.legend(["czerwony", "zielony", "niebieski"])

ax.set_xticks(x)
ax.set_xticklabels(labels)
# ax.legend()

# ax.bar_label(b1, padding=3)
# ax.bar_label(b2, padding=3)
# ax.bar_label(b3, padding=3)

# fig.tight_layout()

plt.show()
```

```python
import pylab

def fun(x, a):
    return round(x * a)


a = 1.7
# b = 2
x = range(0, 256)  # lista argumentów x

y = []  # lista wartości
for i in x:
    val = fun(i - 50, a)
    if val >= 255:
        val = 255
    elif val <= 0:
        val = 0
    y.append(val)

LUT = np.array(y)
# print(y)
pylab.plot(x, LUT, color="red")

pylab.title("Wykres f(x) = a*x + b")
pylab.show()
```
