```python
import numpy as np
m = 9
n = 5
#matrix = np.empty([m, n], dtype=int)
#matrix = np.array([[1, 2], [3, 4], [5, 6]])
#matrix = np.array([[1,2,3,4,5],[2,7,3,4,1],[0,9,7,4,3],[2,3,5,1,2],[7,7,2,1,2],[3,6,7,6,1],[3,4,8,2,9],[9,3,6,8,1],[2,6,8,2,3]])
matrix = np.array(range(m*n),dtype=np.byte)
matrix.shape = (m,n)
print(matrix)
```

```python
#matrix_reshape = matrix.reshape((m,m))
#print(matrix_reshape)
#matrix.resize((m,m),refcheck=False)
#print(matrix)
if m>n:  
    matrix_extend = np.zeros((m,m), dtype=int)
    #print(matrix_resize)

#print(len(matrix[0])) -> 5
#print(len(matrix)) -> 9

start = int((m-n)/2)
#print(start)
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        matrix_extend[i][j+start] = matrix[i][j]
#        print(val)
print(matrix_extend)
```

```python
if m>n:  
    matrix_cut = np.zeros((n,n), dtype=int)
    #print(matrix_cut)

start = int((m-n)/2)
for i in range(n):
    for j in range(n):
        matrix_cut[i][j] = matrix[i+start][j]

print(matrix_cut)
```
